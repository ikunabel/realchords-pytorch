```python
# Raw JSONL passthrough (no parsing, no truncation, no headers)
import itertools
from pathlib import Path
import gzip
import sys

jsonl_path =  '../../data/hooktheory/Hooktheory_Raw.json'
n_lines = 2500  # Set None or <=0 to stream entire file

def open_text(path):
    return gzip.open(path, 'rt', encoding='utf-8') if str(path).endswith('.gz') else open(path, 'rt', encoding='utf-8')

p = Path(jsonl_path)
if not p.exists():
    raise FileNotFoundError(f'File not found: {p}')

with open_text(p) as f:
    it = itertools.islice(f, n_lines) if (n_lines and n_lines > 0) else f
    for line in it:
        # Write exactly as read (no extra newlines, no labels)
        sys.stdout.write(line)
```

```python
import json
from pathlib import Path
from collections import Counter

# Collect interval counts across all datasets
interval_counts = Counter()
datasets = {'wikifonia'}
dataset_results = {}

cache_dir = Path("../../data/cache")

for dataset_name in datasets:
    dataset_path = cache_dir / dataset_name / "train.jsonl"
    
    if not dataset_path.exists():
        print(f"Dataset not found: {dataset_path}")
        continue
    
    song_count = 0
    with open_text(dataset_path) as f:
        for line in f:  # read entire file
            data = json.loads(line)
            harmony = data['annotations']['harmony']
            song_count += 1
            
            for h in harmony:
                intervals = tuple(h['root_position_intervals'])
                interval_counts[intervals] += 1
    
    dataset_results[dataset_name] = song_count
    print(f"{dataset_name}: {song_count} songs")

total_songs = sum(dataset_results.values())
print(f"\n\nTotal songs analyzed: {total_songs}")
print(f"Unique interval patterns found: {len(interval_counts)}")
print("\nAll unique patterns (count):")
for pattern, count in sorted(interval_counts.items()):
    print(f"  {list(pattern)} -> {count}")
```

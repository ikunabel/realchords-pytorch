```python
# Extract song id, song name, and artist from train_set
songs = []
for song in train_set:
    hooktheory = song.get('hooktheory', {})
    song_id = hooktheory.get('id')
    song_name = hooktheory.get('song')
    artist = hooktheory.get('artist')
    songs.append({
        'id': song_id,
        'song': song_name,
        'artist': artist
    })

print(f"Total songs: {len(songs)}")
for song in songs:
    print(f"ID: {song['id']:<20} | Artist: {song['artist']:<30} | Song: {song['song']}")
```

```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine, inspect

# Connect to the wjazzd database
db_path = '../data/wjazzd/wjazzd.db'
engine = create_engine(f'sqlite:///{db_path}')

# Get database inspector to explore schema
inspector = inspect(engine)

# List all tables
tables = inspector.get_table_names()
print("Tables in database:")
for table in tables:
    print(f"  - {table}")
```

```python
# Read all tables and export to CSV
import os
from sqlalchemy import text

# Create csv folder if it doesn't exist
os.makedirs('../data/wjazzd/csv_all', exist_ok=True)

for table_name in tables:
    # Load table data using read_sql
    df = pd.read_sql(f"SELECT * FROM {table_name}", engine)
    
    # Export to CSV
    csv_path = f'../data/wjazzd/csv_all/{table_name}.csv'
    df.to_csv(csv_path, index=False)
    
    print(f"Exported {table_name}: {len(df)} rows, {len(df.columns)} columns → {csv_path}")

print(f"\nAll tables exported to ../data/wjazzd/csv_all/ folder!")
```
